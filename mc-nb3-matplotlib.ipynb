## Welcome to Matplotlib!

Matplotlib is a powerful Python library for creating static, animated, and interactive visualizations.  
In this notebook, you'll learn the basics of Matplotlib and create your own plots interactively.

Let's get started!

```python
# Import Numpy and Matplotlib
import numpy as np
import matplotlib.pyplot as plt

# Below is a basic plot of a line
x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.figure(figsize=(8, 4))
plt.plot(x, y, label='sin(x)')
plt.title('Basic Line Plot')
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.legend()
plt.show()
```

## Jupyter Notebook Widgets

The library ipywidgets, imported below, lets you create interactive controls in Jupyter Notebooks, such as sliders, dropdowns, and buttons. These widgets make it easy to update plots and data dynamically without re-running code.

The code cell below demonstrates how to use ipywidgets to interactively change the frequency and amplitude of a sine wave plot. By moving the sliders, you can see how the graph updates in real time.

```python

%pip install ipywidgets



from IPython.display import display, Markdown
import ipywidgets as widgets


# Interactive Plot with Widgets

def plot_func(freq=1.0, amplitude=1.0):
    plt.figure(figsize=(8, 4))
    y = amplitude * np.sin(freq * x)
    plt.plot(x, y, label=f'{amplitude}*sin({freq}*x)')
    plt.title('Interactive Sine Wave')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.legend()
    plt.show()

widgets.interact(plot_func, freq=(0.5, 5.0, 0.1), amplitude=(0.5, 2.0, 0.1));
```

## Customizing Your Plots

Matplotlib allows you to customize colors, markers, line styles, and more.  
Try changing the parameters in the code below to see different effects!

```python
# Customizing Plots

def custom_plot(color='blue', marker='o', linestyle='-'):
    plt.figure(figsize=(8, 4))
    plt.plot(x, y, color=color, marker=marker, linestyle=linestyle)
    plt.title('Customized Line Plot')
    plt.xlabel('x')
    plt.ylabel('sin(x)')
    plt.show()

widgets.interact(custom_plot, color=['blue', 'green', 'red', 'purple'],
                 marker=['o', 's', '^', 'D'],
                 linestyle=['-', '--', '-.', ':']);
```

## Labels/Graph Keys

Labels in Matplotlib plots are essential for understanding the data being visualized. They help identify what each axis represents and distinguish between multiple datasets or lines within a graph.

- **Axis Labels:**  
    Use `plt.xlabel()` and `plt.ylabel()` to describe the x-axis and y-axis, respectively. This clarifies what values or units are shown.

- **Title:**  
    Add a title to your plot with `plt.title()` to summarize what the graph displays.

- **Legend (Graph Key):**  
    When plotting multiple datasets, use `plt.legend()` to add a key that explains which line or marker corresponds to each dataset.  
    Provide a `label` argument in your plot function (e.g., `plt.plot(x, y, label='sin(x)')`) to specify legend entries.

Clear labels and legends make your plots easier to read and interpret, especially when sharing with others.

```python
# Example: Plotting multiple functions with labels, title, and legend

sin_x = np.sin(x)
cos_x = np.cos(x)

plt.figure(figsize=(8, 4))
plt.plot(x, sin_x, color='blue', linestyle='-', label=r'$\sin x$')
plt.plot(x, cos_x, color='green', linestyle='--', label=r'$\cos x$')


plt.xlabel('x')
plt.ylabel('y')
plt.title('Plot of Multiple Functions')
plt.legend()
plt.show()
```
